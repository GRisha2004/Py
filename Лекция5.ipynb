генерация csv

```python
import csv
import random
from datetime import datetime, timedelta

# Генерация фейковых данных
products = ['Товар A', 'Товар B', 'Товар C']
num_records = 100

with open('data.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['ID продажи', 'Дата продажи', 'Продукт', 'Количество', 'Цена за единицу'])

    for i in range(num_records):
        sale_id = i + 1
        sale_date = datetime.now() - timedelta(days=random.randint(0, 30))
        product = random.choice(products)
        quantity = random.randint(1, 10)
        price_per_unit = round(random.uniform(10, 100), 2)

        writer.writerow([sale_id, sale_date.strftime('%Y-%m-%d'), product, quantity, price_per_unit])
```

```python
import csv
from collections import defaultdict

def calculate_total_income(filename):
    total_income = defaultdict(float)

    with open(filename, mode='r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            product = row['Продукт']
            quantity = int(row['Количество'])
            price_per_unit = float(row['Цена за единицу'])
            income = quantity * price_per_unit
            total_income[product] += income

    return total_income

if __name__ == "__main__":
    filename = 'data.csv'
    total_income = calculate_total_income(filename)

    print("Общий доход по каждому продукту:")
    for product, income in total_income.items():
        print(f"{product}: {income:.2f} руб.")
```

Задача 2

```python
import csv
import json
from collections import defaultdict

def calculate_total_income(filename):
    total_income = defaultdict(float)
    sales_data = []

    with open(filename, mode='r', encoding='utf-8') as file:
        reader = csv.DictReader(file)
        for row in reader:
            sales_data.append(row)  # Сохраняем каждую строку в список
            product = row['Продукт']
            quantity = int(row['Количество'])
            price_per_unit = float(row['Цена за единицу'])
            income = quantity * price_per_unit
            total_income[product] += income

    return total_income, sales_data

def save_to_json(data, filename):
    with open(filename, mode='w', encoding='utf-8') as file:
        json.dump(data, file, ensure_ascii=False, indent=4)

if __name__ == "__main__":
    csv_filename = 'data.csv'
    json_filename = 'data.json'

    total_income, sales_data = calculate_total_income(csv_filename)

    print("Общий доход по каждому продукту:")
    for product, income in total_income.items():
        print(f"{product}: {income:.2f} руб.")

    # Сохранение данных в формате JSON
    save_to_json(sales_data, json_filename)
    print(f"\nДанные успешно сохранены в файл {json_filename}.")
```

```python
import requests

def fetch_posts():
    # Создаем объект Session
    with requests.Session() as session:
        # Отправляем GET-запрос к API
        response = session.get('https://jsonplaceholder.typicode.com/posts')

        # Проверяем, успешен ли запрос
        if response.status_code == 200:
            # Десериализуем ответ в Python-объект
            posts = response.json()
            return posts
        else:
            print(f"Ошибка при получении данных: {response.status_code}")
            return None

if __name__ == "__main__":
    posts = fetch_posts()

    if posts is not None:
        # Выводим все посты в столбик
        for post in posts:
            print(f"ID: {post['id']}")
            print(f"Заголовок: {post['title']}")
            print(f"Содержимое: {post['body']}")
            print("-" * 40)  # Разделитель между постами
```

```python
import requests
from collections import defaultdict

def fetch_posts():
    with requests.Session() as session:
        response = session.get('https://jsonplaceholder.typicode.com/posts')
        if response.status_code == 200:
            return response.json()
        else:
            print(f"Ошибка при получении данных: {response.status_code}")
            return None

def fetch_comments():
    with requests.Session() as session:
        response = session.get('https://jsonplaceholder.typicode.com/comments')
        if response.status_code == 200:
            return response.json()
        else:
            print(f"Ошибка при получении данных: {response.status_code}")
            return None

def find_post_with_max_comments(posts, comments):
    comment_count = defaultdict(int)

    # Подсчитываем количество комментариев для каждого поста
    for comment in comments:
        post_id = comment['postId']
        comment_count[post_id] += 1

    # Находим пост с максимальным количеством комментариев
    max_comments_post_id = max(comment_count, key=comment_count.get)
    max_comments_count = comment_count[max_comments_post_id]

    # Находим сам пост
    max_comments_post = next(post for post in posts if post['id'] == max_comments_post_id)

    return max_comments_post, max_comments_count

if __name__ == "__main__":
    posts = fetch_posts()
    comments = fetch_comments()

    if posts is not None and comments is not None:
        max_comments_post, max_comments_count = find_post_with_max_comments(posts, comments)

        print(f"Пост с максимальным количеством комментариев:")
        print(f"ID: {max_comments_post['id']}")
        print(f"Заголовок: {max_comments_post['title']}")
        print(f"Содержимое: {max_comments_post['body']}")
        print(f"Количество комментариев: {max_comments_count}")
```
